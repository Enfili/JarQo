```python
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import DensityMatrix
from qiskit_aer import Aer
from qiskit.circuit.random import random_circuit

import matplotlib.pyplot as plt
import numpy as np
```

```python
# CONSTANTS
NUMBER_REGISTERS = 2 # data and program
DATA_DIMENSION = 2
PROGRAM_DIMENSION = 2
```

```python
data_register = QuantumRegister(DATA_DIMENSION, 'data')
program_register = QuantumRegister(PROGRAM_DIMENSION, 'program')
quantum_processor = QuantumCircuit(data_register, program_register)

# Generate a random circuit with the same number of qubits and a specified depth
num_qubits = len(data_register) + len(program_register)
depth = 1  # You can choose any depth you prefer
random_qc = random_circuit(num_qubits, depth, max_operands=2)

full = quantum_processor.compose(random_qc)

display(full.draw(output='mpl'))
```
